# Iowa Liquor Sales (Linear Regression)

### Description
The State of Iowa publishes liquor sales data for Iowa Class "E" liquro license, which allows stores to sell liquor.  In this project, I use a simple linear regression to estimate 2016 liquor sales data based on 2015 sales datta.

### Dataset
Iowa liquor sales data can be obtained from [Iowa's Open Data Portal](https://data.iowa.gov/Economy/Iowa-Liquor-Sales/m3tr-qhgy/data).

```python
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn import linear_model
from sklearn.cross_validation import train_test_split
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
## Load the data into a DataFrame
df=pd.read_csv('dataset/Iowa_Liquor_sales_sample_10pct.csv')
print df.columns
```

```python
df.dtypes
```

```python
#rename columns to make them easier to work with
df.columns=[col.replace(' ', '_') for col in df.columns]
df.columns=[col.replace('(', '') for col in df.columns]
df.columns=[col.replace(')', '') for col in df.columns]

## Transform the dates and sale_dollars to datetime and float, respectively
df["Date"] = pd.to_datetime(df["Date"], errors='raise', format="%m/%d/%Y")
df["Sale_Dollars"] = df["Sale_Dollars"].replace('[\$,]','',regex=True).astype(float)
```

```python
#See which columns have missing values
df.isnull().sum()
```

```python
df.shape
```

```python
df.County.isnull().sum()/float(len(df))
```

```python
#drop rows with missing values because they account for a very small fraction of all rows
df.dropna(inplace=True)
```

```python
#create another column that extracts sales data only when sales were done between Jan and Mar of 2015
#this will be used later to compare against Jan-Mar 2016 data, as full year 2016 data is not available
df['Sales_Jan_Mar']=[y if ((x.month >=1) & (x.month <=3)) else 0.0 for x, y in zip(df['Date'],df['Sale_Dollars'])]
```

```python
#statistical description of the dataframe
df.describe()
```

```python
#separate 2015 and 2016 data
df_2015=df[df['Date'].dt.year==2015]
df_2016=df[df['Date'].dt.year==2016]
```

```python
#correlation between features
sns.heatmap(df_2015.corr())
```

```python
import numpy as np
#grouping 2015 data by store
df_2015_groupby=df.groupby(df_2015.Store_Number,as_index=False).agg({
        "Store_Number": lambda x: x.iloc[0],
        "Bottles_Sold": np.sum,
        "Volume_Sold_Gallons": np.sum,
        "Sale_Dollars": np.sum,
        "Sales_Jan_Mar": np.sum,
        "County": lambda x: x.iloc[0]
                })
#grouping 2016 data by store
df_2016_groupby=df.groupby(df_2016.Store_Number,as_index=False).agg({
        "Store_Number": lambda x: x.iloc[0],
        "Bottles_Sold": np.sum,
        "Volume_Sold_Gallons": np.sum,
        "Sale_Dollars": np.sum,
        "Sales_Jan_Mar": np.sum,
        "County": lambda x: x.iloc[0]
                })
df_2015_groupby.head(5)
```

```python
#grouping data by liquor type for visualization
df_top_categories_2015=df.groupby(df_2015.Category_Name,as_index=False).agg({
        "Sale_Dollars": np.sum,
        "Category_Name": lambda x: x.iloc[0]
                }).sort_values(by='Sale_Dollars', ascending=False)
#grouping data by county for visualization
df_top_counties_2015=df.groupby(df_2015.County,as_index=False).agg({
        "Sale_Dollars": np.sum,
        "County": lambda x: x.iloc[0]
                }).sort_values(by='Sale_Dollars', ascending=False)
```

```python
#selecting top categories and counties (by sale dollars) for visualization
df_top_categories_2015=df_top_categories_2015.head(10)
df_top_counties_2015=df_top_counties_2015.head(10)
```

```python
#bar graph showing top 10 liquors sold
plt.figure(figsize=(8,5))
plt.style.use('fivethirtyeight')
ax=sns.barplot(x = 'Sale_Dollars', y = 'Category_Name', data = df_top_categories_2015)
ax.set(ylabel='Liquor Type', xlabel='Sales Amount ($)' , title='Liquor Sales by Type')
```

```python
#bar graph showing top 10 counties with most sales
plt.figure(figsize=(8,4))
plt.style.use('fivethirtyeight')
ax=sns.barplot(x = 'Sale_Dollars', y = 'County', data = df_top_counties_2015)
ax.set(ylabel='County', xlabel='Sales Amount ($)' , title='Liquor Sales by County')
```

```python
#scatter plot to see the correlation between bottles sold, volumes sold, and sales from Jan-Mar vs. full year sale dollars.
sns.set(color_codes=True)
for column in ["Bottles_Sold","Volume_Sold_Gallons","Sales_Jan_Mar"]:
    plt.figure(figsize=(4,4))
    sns.regplot(x=column, y="Sale_Dollars", data=df_2015_groupby)
```

### Linear Regression Model

```python
lm = linear_model.LinearRegression()
from patsy import dmatrices
y_2015,X_2015 = dmatrices("Sale_Dollars ~ Bottles_Sold+Volume_Sold_Gallons+Sales_Jan_Mar+County", df_2015_groupby)
y_2016,X_2016 = dmatrices("Sale_Dollars ~ Bottles_Sold+Volume_Sold_Gallons+Sales_Jan_Mar+County", df_2016_groupby)
```

```python
# Train-Test-Split for cross-validation
from sklearn.cross_validation import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_2015, y_2015, test_size = 0.30)
#predicting 2015 yearly sales using linear regression
lm.fit(X_train, y_train)
predictions = lm.predict(X_test)
print "Train Score: "+str(lm.score(X_train, y_train))
```

```python
model = lm.fit(X_2015, y_2015)
predictions = lm.predict(X_2016)
print "MSE:", mean_squared_error(y_2016, predictions)
print "Model fit:", lm.score(X_2015, y_2015)

#scatter plot to show the correlation between actual 2016 sales vs. predicted 2016 sales
plt.scatter(predictions, y_2016, s=50, c='r', marker='+', zorder=10)
plt.xlabel("Predicted Values")
plt.ylabel("Actual Values")
plt.show()
```

```python
#same graph as above, zoomed into bottom left to see points more closely
fig=plt.figure(figsize=(12,5))
axes=fig.gca()
plt.scatter(predictions, y_2016, s=50, c='steelblue', zorder=10)
axes.set_title('Predicted vs. Actual 2016 Liquor Sales',fontsize=20)
axes.set_xlabel("Predicted Values",fontsize=16)
axes.set_ylabel("Actual Values",fontsize=16)
axes.set_xlim(-10000,150000)
axes.set_ylim(-10000,100000)
plt.show()
```

```python
# Predicted vs. actual
model = lm.fit(X_2015, y_2015)
predictions = lm.predict(X_2015)
predicted_2015 = sum(predictions)

actual_2015 = sum(df_2015['Sale_Dollars'])
actual_2015_q1 = sum(df_2015['Sales_Jan_Mar'])

model = lm.fit(X_2015, y_2015)
predictions = lm.predict(X_2016)
predicted_2016 = sum(predictions)

actual_2016 = sum(df_2016['Sale_Dollars'])

print "Predicted 2015: $"+str(predicted_2015[0])
print "Actual 2015: $"+str(actual_2015)
print "Actual 2015 (Jan-Mar): $"+str(actual_2015_q1)
print "Predicted 2016 (Jan-Mar): $"+str(predicted_2016[0])
print "Actual 2016 (Jan-Mar): $"+str(actual_2016)
```
